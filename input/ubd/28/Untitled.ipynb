```python
import pandas as pd
inputData = pd.read_csv("CH_yearly_air_immission_basetable.csv")

uniqueAggregationIDs = set()
for index, row in inputData.iterrows():
    uniqueAggregationIDs.add(str(str(row['aggregation_id'])+"_"+str(row['unit_id']+"_limit")).replace('/','per'))
    uniqueAggregationIDs.add(str(str(row['aggregation_id'])+"_"+str(row['unit_id']+"_value")).replace('/','per'))

columns = ['station_id', 'pollutant_id', 'year']
for entry in sorted(uniqueAggregationIDs):
    columns.append(str(entry))

outputData = pd.DataFrame(columns)

for index, row in inputData.iterrows():
    new_row = {'station_id':row['station_id']}
    new_row['pollutant_id'] = row['pollutant_id']
    new_row['year'] = int(row['year'])
    new_row[str(str(row['aggregation_id'])+"_"+str(row['unit_id']+"_limit")).replace('/','per')] = row['limitvalue']
    new_row[str(str(row['aggregation_id'])+"_"+str(row['unit_id']+"_value")).replace('/','per')] = row['value']
    outputData = outputData.append(new_row, ignore_index=True)

outputData1 = outputData.groupby(['station_id', 'pollutant_id', 'year']).sum()
outputData = outputData1.head(40000)
outputData.to_csv("output.csv")
```
